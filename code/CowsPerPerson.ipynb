```python
import folium
import pandas as pd

df = pd.read_csv("cows_per_person.csv")

m = folium.Map(location=[52, 5], zoom_start=7)

folium.Choropleth(
    geo_data="the-netherlands.geojson",
    name='choropleth',
    data=df,
    columns=['Regions', 'CowsPerPerson'],
    key_on='feature.properties.name',
    fill_color='BuGn',
    fill_opacity=0.9,
    line_opacity=0.2,
    smooth_factor=0.9,
    legend_name='No of  Dairy Cows (>=2 years) per person for every province'
).add_to(m)

folium.LayerControl().add_to(m)
m

```
